#**INTRODUÇÃO**

A análise a baixo foi feita utilizando dados da American Veterinary Medical Association via data.world disponibilizados no kaggle.com.

A motivação deste projeto é entender a preferência das pessoas na adoção de gatos ou cachorros.



1.   **Quais estados possuem maiores indices de lares com cachorros.**
2.   **Quais estados possuem maiores indices de lares com gatos e cachorros**
3.   **Em termos quantitativos as pessoas tem mais gatos ou chachorros?**




```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cats_vs_dogs.csv", sep =',')
```

```python
df.columns
```

```python
df.columns.drop("Unnamed: 0")
```

```python
df.columns =['nada','estado', 'n_familias', 'porcent_pet_familias', 'familias_possuem_pet', 'porcent_dog_prop', 'familias_dog','dogs_per_household',
       'quantidade_dog', 'percent_prop_gato', 'n_gato_familia',
       'avg_cats_per_household', 'cat_population']
```

```python
dados = df.drop('nada',axis=1)
```

```python
dados.head()
```

**Quantidade de Cachorros com familias**

```python
familias_com_dog_valor = dados[['familias_dog']].sum()
```

```python
familias_com_dog_valor
```

**quantidade total de cachorros**

```python
quant_dog = dados[['quantidade_dog']].sum()
```

```python
quant_dog
```

**Quantidade de gatos com familias**

```python
familias_com_gato = dados[['n_gato_familia']].sum()
```

```python
familias_com_gato
```

**Quantidade de total gato**

```python
gato_total= dados[['cat_population']].sum()
```

```python
gato_total
```

**Quantidade de Cidades analisadas**

**GATOS E CACHORROS COM FAMILIA E SUA RESPECTIVA QUANTIDADE EM CADA ESTADO**

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="familias_dog", y="estado", data=dados,
            label="resultado", color="b")
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="n_gato_familia", y="estado", data=dados,
            label="resultado", color="b")
```
